```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
df=sb.load_dataset('tips')
g1=sb.FacetGrid(df,col="sex",hue="smoker")
g=sb.FacetGrid(df,col="sex",hue="time")
g.map(plt.scatter,"total_bill","tip")
g1.map(plt.scatter,"total_bill","tip")
plt.show()
```

```python
sb.barplot(x="sex",y="tip",hue="time",data=df)
plt.show()
```

```python
sb.regplot(x="total_bill",y="tip",data=df)
plt.show()
```

```python
sb.pairplot(df)
plt.show()
```

```python
selected_vars=['total_bill','tip']
sb.pairplot(df,vars=selected_vars)
plt.show()
```

```python
sb.pairplot(df,hue='size')
plt.show()
```

```python
sb.pairplot(df,hue='size',palette="husl")
plt.show()553
```

```python
sb.pairplot(df,diag_kind='kde')
plt.show()
```

```python
sb.pairplot(df,kind='reg')
plt.show()
```

```python
#controlling the markers
sb.pairplot(df,hue='sex',markers=["o","s"])
plt.show()
```

```python
#advanced customization with seaborn pairplot
g=sb.pairplot(df,hue='day')
g.fig.suptitle("Pairplot of Tips Dataset",y=1.02)
g.set(xticks=[],yticks=[])
plt.show()
```
